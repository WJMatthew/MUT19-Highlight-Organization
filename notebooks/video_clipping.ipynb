```python
from moviepy.video.io.ffmpeg_tools import ffmpeg_extract_subclip

start_time=45
end_time=60
video_file = '/Volumes/PS4 MyPassp/PS4/SHARE/Video Clips/Madden NFL 19/fake/Madden NFL 19_20190518221223.mp4'
ffmpeg_extract_subclip(video_file, start_time, end_time, targetname="test.mp4")
```

```python
import os

video_folder_path = '/Volumes/PS4 MyPassp/PS4/SHARE/Video Clips/Madden NFL 19'

if not os.path.isdir(os.path.join(video_folder_path, 'fake', 'trimmed')):
        os.mkdir(os.path.join(video_folder_path, 'fake', 'trimmed'))
```

```python
fakes = os.listdir(os.path.join(video_folder_path, 'fake'))

for f in fakes:
    try:
        video_file = os.path.join(video_folder_path, 'fake', f)
        ffmpeg_extract_subclip(video_file, start_time, end_time, 
                           targetname=os.path.join(video_folder_path, 'fake', 'trimmed', f))
    except:
        print('ERROR:', f)
```
